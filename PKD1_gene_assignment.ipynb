# BEE-102 Gene Assignment
**Gene:** PKD1 | **Ensembl ID:** ENSG00000160113 | **Tissue:** Kidney | **Function:** Prevents kidney cyst formation

## Setup

```r
library(ensembldb)
library(EnsDb.Hsapiens.v86)

edb <- EnsDb.Hsapiens.v86
```

## Q2 — List all transcripts for PKD1

```r
pkd1_tx <- transcripts(
  edb,
  filter = GeneNameFilter("PKD1"),
  columns = c("tx_id", "tx_biotype", "tx_seq_start", "tx_seq_end", "gene_name", "gene_id")
)

as.data.frame(pkd1_tx)
```

## Q3 — List all proteins for PKD1

```r
pkd1_prot <- proteins(
  edb,
  filter = GeneNameFilter("PKD1"),
  columns = c("protein_id", "tx_id", "gene_name", "gene_id")
)

as.data.frame(pkd1_prot)
```

## Q4 — Protein to genome map for PKD1

```r
pkd1_prot_ids <- proteins(
  edb,
  filter = GeneNameFilter("PKD1"),
  columns = c("protein_id", "protein_sequence")
)

all_maps <- list()

for (pid in pkd1_prot_ids$protein_id) {
  prot_len <- nchar(pkd1_prot_ids[pkd1_prot_ids$protein_id == pid, "protein_sequence"])
  prot_range <- IRanges(start = 1, end = prot_len, names = pid)
  genome_map <- proteinToGenome(prot_range, edb)
  map_df <- as.data.frame(genome_map[[pid]])
  map_df$protein_id <- pid
  all_maps[[pid]] <- map_df
}

pkd1_map <- do.call(rbind, all_maps)
rownames(pkd1_map) <- NULL

write.table(pkd1_map, "PKD1_protein_to_genome_map.tsv", sep = "\t", quote = FALSE, row.names = FALSE)

head(pkd1_map, 20)
```
